```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as dis
import numpy as np
import os, sys
from tqdm import tqdm
import pandas as pd
import librosa.display
```

```python
rawdata = pd.read_csv('Raw_data_plus.csv')
```

```python
# 경로 설정
path = 'C:\\Users\\인공지능사관학교\\Desktop\\최종프로젝트\\data_wav_plus'
mel_path = 'C:\\Users\\인공지능사관학교\\Desktop\\최종프로젝트\\melspectrogram_data_plus'
errors = []

# melspectrogram 을 numpy 배열 형태로 저장
for i in tqdm(range(len(rawdata))):
    if str(rawdata['song_id'].iloc[i])+'.npy' in os.listdir(mel_path):
        continue
    
    music_path = path + '\\' + str(rawdata['song_id'].iloc[i]) + '.wav'
    try:
        
        y, sr = librosa.load(music_path)
    except:
        errors.append(music_path)
        continue
    
    S = librosa.feature.melspectrogram(y=y, sr=sr)
    S_DB = librosa.amplitude_to_db(S, ref=np.max)
    
    np.save('C:\\Users\\인공지능사관학교\\Desktop\\최종프로젝트\\melspectrogram_data_plus\\'+str(rawdata['song_id'].iloc[i])+'.npy',
            S_DB)
    
```
